```python
import matplotlib.pyplot as plt
import csv
plt.rcParams["figure.figsize"] = (22.5, 14.5)
```

```python
x = []
y = []
  
with open('convergence.dat','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    for row in lines:
        x.append(row[0])
        y.append(float(row[1]))
  
plt.plot(x, y, color = 'g', linestyle = 'dashed',
         marker = 'o')
  
# plt.xticks(rotation = 25)
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Generación')
plt.ylabel('Aptitud')
plt.title('Convergencia', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```
